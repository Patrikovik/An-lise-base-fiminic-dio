```python
## QUAIS SÂO OS CRIMES MAIS ESPÉCIES DE CRIMES MAIS COMUNS?

dados_especie = dados_crimes.groupby('ESPECIE')['ESPECIE'].count().sort_values(axis = 0, ascending = False).head(10)


sns.set_style('whitegrid')
plt.figure(figsize=(10,10))
ax = sns.barplot(x = dados_especie.values, y = dados_especie.index,
                 palette = sns.color_palette('Blues',n_colors = len(dados_especie)+4))     
            
plt.show()  
```

```python
dados_rubrica = dados_crimes.groupby('RUBRICA')['RUBRICA'].count().sort_values(axis = 0, ascending = False).head(10)
sns.set_style('whitegrid')
plt.figure(figsize=(10,10))
ax = sns.barplot(x = dados_rubrica.values, y = dados_rubrica.index,
                 palette = sns.color_palette('Blues',n_colors = len(dados_rubrica)+4))     
            
plt.show()  
```

```python
l_ano = []
l_per = []
l_multi = []
l_final = []
l_multi = Ax.tolist()
l_multi2 = []
for ele in anos:
    for ele2 in periodo:
        l_ano.append(ele)
        l_per.append(ele2)
for t in zip(l_ano,l_per):
    l_final.append(t)
for ele3 in l_multi:
        l_multi2.append(ele3*10)
bubble = pd.DataFrame(l_final,columns = ['ANOS','PERIDODO'])
bubble['WEIGHT'] = l_multi
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'ANOS', y= 'PERIDODO', size ='WEIGHT',  data = bubble )
plt.show()
```
